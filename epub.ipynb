```python
import ebooklib
from ebooklib import epub
```

```python
EBOOK_PATH = "data/epubs/Sorceleur - L'Integrale - Andrzej Sapkowski.epub"
```

```python
book = epub.read_epub(EBOOK_PATH)
print(book.get_metadata('DC', 'title'))
print(book.get_metadata('DC', 'language'))
print(book.get_metadata('DC', 'creator'))
print(book)
```

```python
documents = list(book.get_items_of_type(ebooklib.ITEM_DOCUMENT))
images = list(book.get_items_of_type(ebooklib.ITEM_IMAGE))
navigation = list(book.get_items_of_type(ebooklib.ITEM_NAVIGATION))
covers = list(book.get_items_of_type(ebooklib.ITEM_COVER))
```

```python
for doc in documents:
    print(doc.get_name(), ' : ', len(doc.get_body_content()))
```

```python
navs = []
for nav in navigation:
    print(nav.book)
    print(nav.content)
    print(nav.file_name)
    print(nav.get_id())
    print(nav.get_name())
    print(nav.get_type()) # 4 = navigation
    print(nav.is_linear)
    navs.append(nav)
```

```python
nav_html = navs[0].content.decode('utf-8')
```

```python
from bs4 import BeautifulSoup
import json

table_of_content = []

nav_soup = BeautifulSoup(nav_html, features="xml")
navmap = nav_soup.navMap

def process_navpoint(navpoint):
    result = {}
    children_navpoints = navpoint.find_all('navPoint')
    if len(children_navpoints) > 0:
        result['children'] = [process_navpoint(child) for child in children_navpoints]
    else:
        result['children'] = []
    
    result['id'] = navpoint['id']
    result['playorder'] = navpoint['playOrder']
    result['label'] = navpoint.navLabel.text
    result['content'] = navpoint.content['src']

    return result

for child in navmap.children:
    if child.name == 'navPoint':
        table_of_content.append(process_navpoint(child))

print(json.dumps(table_of_content, indent=4))

```
