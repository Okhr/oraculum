```python
navs = []
for nav in navigation:
    print(nav.book)
    print(nav.content)
    print(nav.file_name)
    print(nav.get_id())
    print(nav.get_name())
    print(nav.get_type()) # 4 = navigation
    print(nav.is_linear)
    navs.append(nav)
```

```python
nav_html = navs[0].content.decode('utf-8')
```

```python
from bs4 import BeautifulSoup
import json

table_of_content = []

nav_soup = BeautifulSoup(nav_html, features="xml")
navmap = nav_soup.navMap

def parse_item(item_href):
    item = book.get_item_with_href(item_href)
    item_body_content = item.get_body_content()
    item_soup = BeautifulSoup(item_body_content)
    # print(item_soup.prettify())
    for p in item_soup.find_all('p'):
        print(p.string)

def get_structured_book_data(navpoint):
    result = {}
    children_navpoints = navpoint.find_all('navPoint')
    if len(children_navpoints) > 0:
        result['children'] = [get_structured_book_data(child) for child in children_navpoints]
    else:
        result['children'] = []
    
    result['id'] = navpoint['id']
    result['playorder'] = navpoint['playOrder']
    result['label'] = navpoint.navLabel.text.strip()
    result['content_path'] = navpoint.content['src']

    return result


for child in navmap.children:
    if child.name == 'navPoint':
        table_of_content.append(get_structured_book_data(child))

#print(json.dumps(table_of_content, indent=4))

parse_item("Sapkowski, Andrzej - Le Dernier Voeu_split_004.htm")
```
